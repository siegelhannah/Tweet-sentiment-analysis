# Part 2: Sentiment Analysis

We can use the words in our Game of Thrones tweets to calculate a measure of the sentiment of the tweet. For example, the sentence "I loved the episode!" has positive sentiment, whereas the sentence "I hate Cersei!" has a negative sentiment. In addition, some words have stronger positive / negative sentiment than others: "I loved the episode!" is more positive than "I liked the episode."

We will use the [VADER (Valence Aware Dictionary and sEntiment Reasoner)](https://github.com/cjhutto/vaderSentiment) lexicon to analyze the sentiment of our tweets. VADER is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media which is great for our usage.

The VADER lexicon gives the sentiment of individual words. Run the following cell to show the first few rows of the lexicon:

```python
print(''.join(open("vader_lexicon.txt").readlines()[:10]))
```

As you can see, the lexicon contains emojis too! Each row contains a word and the *polarity* of that word, measuring how positive or negative the word is.

The creators of VADER describe the tool’s assessment of polarity, or “compound score,” in the following way:

“The compound score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. Calling it a 'normalized, weighted composite score' is accurate.”

As you can see, VADER doesn't "read" sentences, but works by parsing sentences into words assigning a preset generalized score from their testing sets to each word separately. 

VADER relies on humans to stabilize its scoring. The creators use Amazon Mechanical Turk, a crowdsourcing survey platform, to train its model. Its training set of data consists of a small corpus of tweets, New York Times editorials and news articles, Rotten Tomatoes reviews, and Amazon product reviews, tokenized using the natural language toolkit (NLTK). Each word in each dataset was reviewed and rated by at least 20 trained individuals who had signed up to work on these tasks through Mechanical Turk. 

### 2.1 - Considering VADER's accuracy in different scenarios

VADER aggregates the sentiment of words in order to determine the overall sentiment of a sentence, and further aggregates sentences to assign just one aggregated score to a whole tweet or collection of tweets. This is a complex process and if you'd like to learn more about how VADER aggregates sentiment, here is the info at this [link](https://github.com/cjhutto/vaderSentiment#introduction).

However, there are circumstances when you might not want to use VADER, like if someone is using sarcasm in their language. This might look like very positive words like "great" being used to convey an actually negative message, which wouldn't be processed correctly by VADER. In this way, it is always important to consider that there may be some features of human speech that VADER could misrepresent or fail to capture.

### 2.2 Converting the lexicon to DataFrame
(Below)

```python
words = []
polarities = []

with open("vader_lexicon.txt", "r") as lexicon:
    for line in lexicon:
        Lsplit = line.strip().split('\t') # split line by item
        words.append(Lsplit[0]) # append word
        polarities.append(float(Lsplit[1])) # append polarity

# make DataFrame
sent = pd.DataFrame({"word": words, "polarity": polarities})
sent
```

### 2.3 - Calculating the sentiment of tweets

Two general steps:
1. For each tweet, we find the sentiment of each word.
2. We calculate the sentiment of each tweet by taking the sum of the sentiments of its words.

#### 2.3.1 

First, we need to standardize the text. There are a number of elements in each tweet that we don't want in our sentiment analysis, like "@" or "#", so we remove these first.

```python
pat = '[@#]' # regular expression pattern

# replace unwanted chars
got["text"] = got["text"].replace(pat, '', regex=True)
got.head()
```

#### 2.3.2

Notice how nearly all of our tweets end with a url. These characters similarly represent nonsense as far as our sentiment analysis is concerned. Our next step therefore should be to further clean up the `text` field by removing URLs from each `text` entry.

```python
pat = 'https.*' # https and anything after
got["text"] = got["text"].replace(pat, '', regex=True)
got.head()
```

#### 2.3.3

Next, let's remove any unusual characters (which generally represent emojis on twitter) and punctuation that won't be useful in our sentiment analysis. *i.e.* Let's only keep upper-case and lower-case alphabetic characters, and spaces. Finally, we lowercase all of the text in the tweets since the lexicon is also lowercase.

```python
pat = '[^A-Za-z ]' # any character NOT upper/lowercase alpha & spaces

got["text"] = got["text"].replace(pat, '', regex=True)

got['text'] = got['text'].str.lower() # lowercase
got.head()
```

### 2.4 - Splitting tweets up into individual words

Now, let's convert the tweets into a tidy format to make the sentiments easier to calculate. We can create a new Dataframe called `tidy_got`, with the following columns:

1. `id`: The id of the tweet, which can take on any value 0 - n unique tweets.
2. `num`: The location of the word in the tweet. For example, if the tweet was "i love america", then the location of the word "i" is 0, "love" is 1, and "america" is 2.
3. `word`: The individual words of each tweet.

```python
select_cols = got[["status_id", "text"]].rename(columns={"status_id": "id"})
expanded = select_cols["text"].str.split(expand=True) # expands every word in each tweet into its own entry in a df

# put all words into a single column, labeled by their position & tweet id
stacked = expanded.stack()
stacked.index.names = ["id", "num"]

stacked_df = stacked.to_frame(name="word") # convert Series to dataframe
reset = stacked_df.reset_index(level=["num"]) # reset index to column
tidy_got = reset
tidy_got.head()
```

### 2.5 - Finding the sentiment

Now that we have this table in the tidy format, it becomes much easier to find the sentiment of each tweet: we can join the table with the lexicon table. 

To do this, we add a `polarity` column to the `got` table.  The `polarity` column contains the sum of the sentiment polarity of each word in the text of the tweet.

```python
sent_unique = sent.drop_duplicates(subset="word")

merged = tidy_got.merge(sent_unique, on = "word", how='left')
merged.index = tidy_got.index # reset index to 0,0,0,0,...1,1,1,...
merged["polarity"] = merged["polarity"].fillna(0) # 0 instead of NaN
grouped = merged.groupby("id").sum()

polarities = grouped["polarity"]
got["polarity"] = polarities
got
```

Now we have a measure of the sentiment of each of his tweets! Note that this calculation is rather basic; you can read over the VADER readme to understand a more robust sentiment analysis.

Now, run the cells below to see the most positive and most negative tweets about Game of Thrones in our dataset:

```python
print('Most negative tweets:')
for t in got.sort_values('polarity').head()['text']:
    print('\n  ', t)
```

```python
print('Most positive tweets:')
for t in got.sort_values('polarity', ascending=False).head()['text']:
    print('\n  ', t)
```

### 2.6 - Analysis

The 5 most negative tweets seem to truly be the most negative, as most of them have a lot of instances of the word "kill" and are generally negatively talking about parts of the show. The 5 most positive tweets seem to be truly positive as well, except for the second one about the "great great great grandson", in which VADER seems to have misinterpreted "great" meaning "good" / being positive rather than just describing family.

### 2.7 - Character sentiment analysis

Now, let's try looking at the distributions of sentiments for tweets containing the names of certain characters. This is shown below in an overlapping KDE plot of tweet sentiments regarding three characters from Game of Thrones: 'daenerys', 'john snow', and 'cersei'.

```python
daenerys = got[got["text"].str.contains("daenerys", case=False, na=False)]
john_snow = got[got["text"].str.contains("john snow", case=False, na=False)]
cersei = got[got["text"].str.contains("cersei", case=False, na=False)]

#PLOT
plt.figure(figsize=(10,8))
sns.histplot(daenerys, x = daenerys["polarity"], kde=True, alpha=0.3, label='daenerys', color='green')
sns.histplot(john_snow, x = john_snow["polarity"], kde=True, alpha=0.3, label='john snow', color='red')
sns.histplot(cersei, x = cersei["polarity"], kde=True, alpha=0.3, label='cersei', color='blue')

plt.legend()
plt.title("tweets about characters & their polarity")
plt.xlabel("tweet polarity")
plt.xlim(-15, 15)
plt.ylim(0,890) 
```

### 2.8 - Conclusion

All three "peaks" for the 3 characters are around 0, with a fairly even distribution on either side. This indicates that the most tweets about each of the characters are neutral in sentiment. There are more total tweets about Cersei than John Snow, and the most tweets about Daenerys. For Daenerys, there seems to be a slightly larger number of negatively-skewed tweets than positive ones. For Cersei, there is also a very slight skew towards negative tweets, but a small peak on the very-positively skewed side, indicating that although most tweets about her are neutral and slightly negative, there are a few that are very positive. For John snow, there are not many tweets at all and range around 0 in terms of polarity/sentiment.
